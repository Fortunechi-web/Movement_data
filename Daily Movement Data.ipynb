# Trend of steps taken over time

```python
# Create the lineplot visualization
plt.figure(figsize=(10, 5))
sns.lineplot(data=daily_movement_data, x='Date (Timestamp)', y='Steps', marker='o')

# Add titles and labels
plt.title('The Trend of Steps Taken Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Steps', fontsize=14)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

### The number of steps taken varies over time, as seen in this line chart from September 17th, 2024, to September 29th, 2024. The date is shown on the x-axis, and the number of steps taken is shown on the y-axis.

### There was a noticeable dip in activity around September 19th–22nd, where the number of steps taken was very low, approaching zero, which had 443, 100, and 236 moves, respectively. After that, the 22nd sees a significant increase, and the steps keep fluctuating with multiple peaks, particularly on the 26th and 28th of September.

### This graph makes it easier to see daily movement patterns and the degree of activity or inactivity on different days.


# Comparison of Distance Traveled and Steps Taken

```python
# Create the scatterplot visualization
plt.figure(figsize=(10, 5))
sns.scatterplot(data=daily_movement_data, x='Steps', y='Distance (Miles)', hue='Flights Covered',
                size='Move (Calories)', sizes=(20, 200))

# Add titles and labels
plt.title('Steps vs Distance Traveled', fontsize=16)
plt.xlabel('Steps', fontsize=14)
plt.ylabel('Distance (Miles)', fontsize=14)

# Show the plot
plt.show()
```

### The steps taken are shown on the x-axis, and the miles traveled are shown on the y-axis. Every data point is represented by a bubble; the color of the bubble indicates the number of flights covered and the size of the bubble indicates the number of calories burned.

### It makes sense that as the number of steps increases, so does the distance traveled in miles. The size of the bubbles indicates that as steps and distance increase, calories burned (represented by larger bubbles) increase as well. Darker bubbles represent more flights covered, indicating more steps or higher physical activity.

### All things considered, the visual shows that there is a clear correlation between the number of steps taken and the distance traveled and that the total number of stairs climbed as well as the activity's intensity affect the amount of calories burned.

# The Distribution of Calories Burned

```python
# Create the histogram visualization
plt.figure(figsize=(10, 5))
sns.histplot(daily_movement_data['Move (Calories)'], bins=10, kde=True, color='purple')

# Add titles and labels
plt.title('The Distribution of Calories Burned', fontsize=16)
plt.xlabel('Calories Burned', fontsize=14)

# Show the plot
plt.show()
```

### The distribution of calories burned is displayed visually. It is a histogram with a superimposed density curve. The histogram's shape is roughly represented by the smooth line known as the superimposed density curve. It gives the data distribution a continuous representation.

### The number of calories burned is shown on the x-axis, and the count or frequency of observations within each range of calories burned is shown on the y-axis. The number of data points falling within a given range is indicated by the height of each bar.

### Significant peaks can be found at calorie-burning ranges of 0–20 and 100–140, respectively. Calorie burn rates in the middle (between 40 and 80) are relatively low. The distribution's peak falls between 0 and 20 calories burned, indicating that this is the most common range.

### There is a skewed distribution of burned calories towards the right. This suggests that there is a distributional tail to the right, meaning that on some days, I burned a lot more calories than on others. This might be the result of things like changing daily routines or engaging in more physical activity.


# Correlation between Steps and Calories Burned

```python
# Create the scatterplot visualization
plt.figure(figsize=(10, 5))
sns.scatterplot(data=daily_movement_data, x='Steps', y='Move (Calories)', color='purple')

# Add titles and labels
plt.title('The Correlation Between Steps and Calories Burned', fontsize=16)
plt.xlabel('Steps', fontsize=14)
plt.ylabel('Calories Burned', fontsize=14)

# Show the plot
plt.show()
```

### The correlation between the number of steps taken and the number of calories burned is shown in this scatter plot. The steps taken are shown on the x-axis, and the calories burned are shown on the y-axis.

### Steps and calories burned have a strong positive correlation, which means that as steps taken increase, so do calories burned. The data points show a steady trend with no significant deviations, confirming that taking more steps each day results in a steady burn of calories.

### The graph does a wonderful job of showing how increasing physical activity (as indicated by steps) increases caloric expenditure.


# Days with Highest Move (Calories)

```python
# Split timestamp into column 'Day'
daily_movement_data["Day"] = daily_movement_data["Date (Timestamp)"].dt.strftime('%A')

# Top 5 days with highest 'Move (Calories)'
top_move_days = daily_movement_data.nlargest(5, 'Move (Calories)')

# Set the size of the plot
plt.figure(figsize=(12, 6))

# Bar chart for Top 5 days with highest 'Move (Calories)'
ax = sns.barplot(data=top_move_days, x='Day', y='Move (Calories)', color='purple')
plt.title('Top Days with Highest Move (Calories)', fontsize=16)
plt.xlabel('Day', fontsize=14)
plt.ylabel('Move (Calories)', fontsize=14)

# Adjust layout
plt.tight_layout()

# Annotate the bars with the total Move (Calories) values
for bar in ax.patches:
    ax.annotate(f'{bar.get_height():,.2f}', 
                (bar.get_x() + bar.get_width() / 2, bar.get_height()),  # position annotation on top of bars
                ha='center', va='center', size=12, xytext=(0, 8), textcoords='offset points')

# Display the plot
plt.show()
```

### The bar chart shows the top days with the highest move (calories). The x-axis displays the days of the week, and the y-axis displays the movement (calories burned).

### Here are the days that I burnt the most calories, starting with Thursday (147 calories) and followed by Saturday (137 calories).

### These days appear to be fairly consistent in terms of activity; however, Sunday and Monday show marginally lower calorie burn than Thursday.


# Days with Highest Steps

```python
# Days with highest 'Steps'
top_steps_days = daily_movement_data.nlargest(5, 'Steps')

# Set the size of the plot
plt.figure(figsize=(12, 6))

# Bar chart for Top 5 days with highest 'Steps'
ax = sns.barplot(data=top_steps_days, x='Day', y='Steps', color='purple')

# Add title and labels
plt.title('Days with the Highest Steps', fontsize=16)
plt.xlabel('Day', fontsize=14)
plt.ylabel('Steps', fontsize=14)

# Adjust layout
plt.tight_layout()

# Annotate the bars with the total Steps values
for bar in ax.patches:
    ax.annotate(f'{bar.get_height():,.0f}',  # No decimals for steps, assuming integer values
                (bar.get_x() + bar.get_width() / 2, bar.get_height()),  # position annotation on top of bars
                ha='center', va='center', size=12, xytext=(0, 8), textcoords='offset points')

# Display the plot
plt.show()
```

### The days with the highest steps are displayed in the bar chart. The days of the week are shown on the x-axis, and the steps taken are shown on the y-axis.

### The day with the most steps was Saturday (5,499), followed by Thursday (5,333). The least amount of steps were taken on Tuesday (4,578). On Sunday, Monday, and Tuesday, the total number of steps was comparatively comparable. 

### This visual indicates that, in comparison to the workday, I am more active on the weekends (Saturday and Sunday).
